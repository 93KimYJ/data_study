```python
rows = xmlobj.findAll(name="item")

rowList = [] 
columnList = []

# 전체 row 데이터 수집
for j in range(0, len(rows)):

    # 첫번째 행의 모든 컬럼 값을 다 찾는다
    columns = rows[j].findAll()
    # 전체 컬럼 데이터 수집
    for i in range(0, len(columns)):

        columnValue = columns[i].text
        columnList.append(columnValue)

    rowList.append(columnList)

    columnList = []

pd.DataFrame(rowList)
```
