### 제품을 4개 이상 구매했고 2년 미만 지속 사용한 고객 조회

```python
countCond = 4
ageCond = 2
```

```python
refinedData = customerData.loc[ (customerData.DEVICECOUNT >= countCond) &
                  (customerData.PRODUCTAGE < ageCond) ]
```

```python
refinedData.loc[ refinedData.PRODUCTAGE >= 2 ]  ## 반대사이즈 검색해서 확인
```

```python
refinedData.AVGPRICE = refinedData.AVGPRICE.astype(str)
```

```python
refinedData.dtypes
```

```python
refinedData.loc [ refinedData.AVGPRICE.astype(float) > 100 ]
```

```python
emiTargets = [2,3,5]
refinedData.loc [ ~refinedData.EMI.isin(emiTargets) ]
```
